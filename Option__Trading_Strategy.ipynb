## 3) Plotting for Multiple Options Strategy

### 3.1) Short Strangle

To set up a Long Strangle, all we need to do is buy a Call with strike K2 and a Put with strike K1. This allows us to benefit from variations in the underlying in the event of a major rise or fall. This means that we expect the underlying asset to be highly volatile. The trader who sells a strangle expects the price of the asset to remain within a certain price range (defined by the put and call strikes). Selling a strangle is therefore a strategy that relies on low volatility, where the maximum gain is limited to the amount of the premium received for selling the options. However, selling a strangle carries unlimited risk if the market goes against your forecasts. If the price of the underlying asset moves significantly in any direction, the losses can be substantial.

```python
op_1 = {'op_type':'c','strike':120,'tr_type':'s','op_pr':2 }
op_2 = {'op_type':'p','strike':100,'tr_type':'s','op_pr':3}
op.multi_plotter(spot=110, op_list=[op_1, op_2])
```

### 3.2) Iron Condor
Being short iron Condor means being better hedged than being short strangle. As well as selling a strangle (call option and put option), the trader also buys a strangle at a higher strike price for the call and a lower strike price for the put. The effect of this additional step is to limit the risk of losses in the event of significant market movements, at the expense of a reduction in the net premium received. As a result, the iron condor has a predefined and limited maximum loss, unlike a short strangle.


```python
op_1 = {'op_type':'c', 'tr_type':'s', 'strike':120, 'op_pr':2}
op_2 = {'op_type':'c', 'tr_type':'b', 'strike':130, 'op_pr':1}
op_3 = {'op_type':'p', 'tr_type':'s', 'strike':100, 'op_pr':3}
op_4 = {'op_type':'p', 'tr_type':'b', 'strike':90, 'op_pr':1}

op.multi_plotter(spot=110, op_list=[op_1,op_2,op_3,op_4])
```
